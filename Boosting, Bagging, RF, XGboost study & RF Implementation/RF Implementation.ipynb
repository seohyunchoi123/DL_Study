```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.cross_validation import train_test_split
from sklearn import metrics
```

```python
cancer = load_breast_cancer()
```

```python
pd.Series(cancer.target).unique()
```

```python
dat = pd.concat([pd.DataFrame(cancer.data, columns= cancer.feature_names), pd.DataFrame(cancer.target, columns=["is_cancer"])], axis=1)
dat 
```

```python
test, train = train_test_split(dat, test_size = 0.7, random_state = 930904)
```

```python
test
```

### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
features_list = dat.columns
```

```python
?RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=10, max_features=6)
rf.fit(X=train[features_list[:-1]], y= train[features_list[-1]])
pred = rf.predict(test[features_list[:-1]])
print(metrics.classification_report(pred, test[features_list[-1]]))
```

```python
def my_rf(train, test, n_estimators, max_features):
    total_votes = pd.DataFrame()
    features_list = train.columns
    for i in range(n_estimators):
        smp_features = np.random.choice(features_list[:-1], max_features, replace=False)
        smp_data_idx =  np.random.choice(list(range(train.shape[0])), train.shape[0], replace=True)
        smp_train = train[smp_features].iloc[smp_data_idx] #  +    1 
        smp_test = test[smp_features]
        
        tree = DecisionTreeClassifier()
        tree.fit(X=smp_train, y= train[features_list[-1]].iloc[smp_data_idx]) #   1  
        
        pred = tree.predict(smp_test) #  1    ! 
        voting = pd.DataFrame({str(i+1) + ' ' : pred}) 
        total_votes = pd.concat([total_votes, voting], axis=1) #    
    sum_of_votes = total_votes.sum(axis=1)
    result = sum_of_votes.apply(lambda x : 1 if x>5 else 0)
    return(list(result))
```

```python
my_pred = my_rf(train, test, n_estimators=10, max_features=6)
print(metrics.classification_report(my_pred, test[features_list[-1]]))
```
